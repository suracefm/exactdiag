```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
import time
from operators import *
```

```python
h=np.pi*0.99
m=1.0
J=1.0
hz=0.1
hx=0.1
n_dis=100
L=2
#index=1
f=0.8
```

```python
sz = np.array([[1,0],[0,-1]])
sx = np.array([[0,1],[1,0]])
```

```python
kick = la.expm(-1j*h*sx/2)
```

```python
def build_H(J,m,hz,hx,L):
    dim = 2**L
    H = np.zeros((dim, dim), dtype=float)
    g = np.random.choice([1,-1], size=L)
    for site in range(L):
        H=H+(J-g[site]*m/2)*two_site_op(sz, sz, site, site+1, L)
        H=H+hx*one_site_op(sx, site, L)
        H=H+hz*one_site_op(sz, site, L)
    return H
```

```python
sizes = [2,3,4,5,6]
time_set = np.power(2, np.arange(40))
Ztot=np.zeros((len(sizes), len(time_set)), dtype=complex)

for countL, L in enumerate(sizes):
    initial_state=np.zeros(2**L)
    flips = L-int(L*f)
    index=np.sum(np.power(2, np.random.randint(L, size=flips)))
    initial_state[index]=1
    print("Initial state (local magnetization) ",[expect_val(one_site_op(sz, i, L), initial_state) for i in range(L)])

    K = kick
    for _ in range(L-1):
        K = np.kron(K, kick)

    
    for counter in range(n_dis):
        start = time.time()

        H = build_H(J,m,hz,hx,L)
        U_F = np.dot(la.expm(-1j*H),K)
        eigval, eigvec = np.linalg.eig(U_F)
        spectrum = np.angle(eigval)

        final_state = evolve(time_set, initial_state, eigvec, eigval)
        _Z=0
        for i in range(L):
            _Z+=expect_val(one_site_op(sz, i, L), initial_state)*\
                     expect_val(one_site_op(sz, i, L),final_state)\
                     *(-1)**(time_set%2)
        Ztot[countL]+=_Z/L
        elapsed = time.time()-start
        #print('size', L, '\tdisorder realization', counter,'\ttime elapsed', elapsed)

Z=Ztot/n_dis

```

```python
for i, row in enumerate(Z):
    plt.plot(np.log10(time_set), row, label='L=%d' %sizes[i])
plt.ylabel(r'$Z(nT)$')
plt.xlabel(r'$\log_{10}(n)$')
plt.legend()
plt.show()
```

```python
n_star = np.zeros(len(sizes), dtype=int32)
for i, row in enumerate(Z):
    n_star[i] = np.argmax(row<0)
plt.plot(sizes, np.log10(time_set[n_star]))
plt.show()
```
