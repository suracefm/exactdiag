```python
import numpy as np
import matplotlib.pyplot as plt
from spectrum_analysis import * 
```

```python
samples=100
logsh=np.empty(samples)
log=np.empty(samples)
r=np.empty(samples)

f, axarr = plt.subplots(4, figsize=(6,12))
lsharr=[]
larr=[]
rarr=[]
sizes=list(range(2,12))
for L in range(2,12):
    for i in range(samples):
        a=2*np.pi*np.random.random_sample(3**L)
        log[i]=np.mean(np.log10(gap(a)))
        logsh[i]=np.mean(np.log10(shifted_gap(a,3)))
        r[i]=ratio(a)
    axarr[0].hist(log, color='C%d' %(L-2), alpha=0.4)
    axarr[0].axvline(x=np.mean(log), color='C%d' %(L-2))
    axarr[1].hist(logsh, color='C%d' %(L-2), alpha=0.4)
    axarr[1].axvline(x=np.mean(logsh), color='C%d' %(L-2))
    axarr[2].hist(r, color='C%d' %(L-2), alpha=0.4)
    axarr[2].axvline(x=np.mean(r), color='C%d' %(L-2))
    lsharr.append(np.mean(logsh))
    larr.append(np.mean(log))
    rarr.append(r)
    axarr[3].plot(1/L, np.mean(logsh)-np.mean(log), marker='o', color='k')
    
plt.show()
```

```python
for L in range(12,15):
    for i in range(samples):
        a=2*np.pi*np.random.random_sample(3**L)
        log[i]=np.mean(np.log10(gap(a)))
        logsh[i]=np.mean(np.log10(shifted_gap(a,3)))
        r[i]=ratio(a)
    lsharr.append(np.mean(logsh))
    larr.append(np.mean(log))
    rarr.append(np.mean(r))
```

```python
for L in range(12,15): sizes.append(L)
```

```python
f, axarr = plt.subplots(4, figsize=(6,12))
axarr[0].plot(1/np.array(sizes), larr)
axarr[1].plot(1/np.array(sizes), lsharr)
axarr[2].plot(1/np.array(sizes), rarr)
axarr[3].plot(1/np.array(sizes), np.array(lsharr)-np.array(larr))
```
